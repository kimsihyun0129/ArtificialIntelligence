# 데이터 다루기

## numpy

### 넘파이의 array() 메소드를 사용하여 배열을 생성

```python
import numpy as np  # numpy 패키지를 np라는 이름으로 불러옴

ar1 = np.array([1,2,3,4,5])
ar2 = np.array([[1,2,3],[4,5,6]])
print(ar1)
print(ar2)
```

### 난수를 이용한 2차원 배열 객체 생성

```python
ar3 = np.random.randn(3) # 3 크기의 1차원 배열 생성(난수 발생)
ar4 = np.random.randn(2,3) # 2*3 크기의 2차원 배열 생성(난수 발생)
print(ar3)
print(ar4)
```

### 원소를 0 또는 1로 초기화

```python
ar5 = np.zeros(5)
ar6 = np.ones((2,3))
print(ar5)
print(ar6)
```

### arange()와 reshape() 메소드 이용

```python
ar7 = np.arange(20,200,10) # 20부터 200-1까지 10씩 증가
ar8 = ar7.reshape(3,6) # 3*6배열로 재배치
print(ar7)
print(ar8)
```

### 배열 다루기

```python
ar1 = np.arange(1,21,1)
print(ar1)
print(ar1[1])

ar2 = ar1.reshape(2,10)
print(ar2)
ar2[1][1] = 100
print(ar2[1][0], ar2[1][1], ar2[1][2])
```

### 배열의 산술 연산

```python
ar1 = np.arange(1,11,1)
ar2 = ar1+3
ar3 = ar1*2

print(ar1)
print(ar2,"\n",ar3)
```

### 배열의 통계 메소드 사용

```python
ar1 = np.array([[5,7,9],[-7,-6,19],[6,9,11]])
print(ar1)
print(ar1.sum())  # 배열 원소의 합계
print(ar1.mean()) # 배열 원소의 평균
print(ar1.max())  # 배열 원소 중 가장 큰 값
print(ar1.min())  # 배열 원소 중 가장 작은 값
print(ar1.max(axis=0))  # 배열의 각 열의 원소 중 가장 큰 값을 구함
print(ar1.max(axis=1))  # 배열의 각 행의 원소 중 가장 큰 값을 구함
```

### 심화 조건식을 사용한 연산

```python
ar1 = np.array([[5,7,9],[-7,-6,19],[6,9,11]])
print(ar1 > 0)  # ar1의 각 원소가 0보다 큰지 판단한 결과를 출력

m_count = (ar1 < 0).sum() # 원소 중 음수가 몇 개인지 계산(false의 개수)
print(m_count)

ar2 = np.where(ar1 < 0, 0, ar1) # ar1의 원소가 0보다 작으면 0으로 변경, 그렇지 않으면 ar1의 값으로 대
print(ar2)
```

## Pandas

### 시리즈(series) 객체 생성하기

```python
import pandas as pd # 판다스 패키지를 불러와서 pd라는 이름으로 사용

sr1 = pd.Series([10,20,30,40,50])  # 리스트를 시리즈 객체로 생성
print(sr1)
```

### 시리즈 객체의 값과 인덱스 확인

```python
sr1 = pd.Series([10,30,20,40,50])

print(sr1.values) # 시리즈 sr1의 값 출력
print(sr1.index)  # 시리즈 sr1의 인덱스 출력(처음값, 마지막값, 증가값)
print(sr1[2])     # 인덱스 2에 해당하는 값 출력
```

### 시리즈(series) 객체의 인덱스 생성

```python
sr1 = pd.Series([10,30,20,40,50], index=['a','b','c','d','e'])

print(sr1[1])     # 숫자 인덱스
print(sr1['b'])   # 문자 인덱스
```

```python
sr1 = pd.Series([10,30,20,40,50], index=['a','b','c','d','e'])

print(sr1[[1,2]])
print(sr1[['a','c']])
print(sr1[1:4])
print(sr1['b':'d'])
```

## 데이터 전처리

### 파생변수 생성

```python
df["합계"] = df["중간"] + df["기말"]  # 중간과 기말 점수를 더해 '합계'열로 생성
x = ["1반","1반","1반","2반","2반","2반","3반","3반","3반","3반"] # 리스트 정의

df["cls"] = x

df
```

### 표준화

```python
mid_avg = np.mean(df["중간"]) # 중간시험 점수 평균
mid_std = np.std(df["중간"])  # 중간시험 점수 표준편차

# 중간시험 Z-score 열 생성
df["중간_Z점수"] = (df["중간"]-mid_avg)/mid_std

df.head()
```

### 정규화

```python
fin_min = np.min(df["기말"])  # 기말 점수 최소값
fin_max = np.max(df["기말"])  # 기말 점수 최대값

# 기말 점수 정규화
df["기말_정규화"] = (df["기말"]-fin_min)/(fin_max-fin_min)

df.head()
```

### 결측값 식별

```python
x = [14,15,13,14,None,None,19,11,12,18]
df["토론"] = x  # 결측값이 있는 컬럼 "토론" 추가

df.info() # 데이터 프레임 요약정보 출력 : NaN의 개수를 알 수 있음
```

### 결측값 대치

```python
# 완전분석법(결측값이 있는 행 제거)
df_new = df.dropna(axis=0)  # 결측값이 있는 행 전체 제거 (axis=1이면 열을 제거)
df_new = df["토론"].dropna() # "토론"열에서 결측값 제거
df_new
```

```python
# 평균대치법(결측값을 평균값으로 대체)
d_mean = df["토론"].mean()
df["토론"].fillna(d_mean, inplace=True) # 결측값을 평균값으로 대치
df
```

### 이상값(outlier) 처리

```python
df_10R = df["기말"].quantile(0.25)  # 1사분위수
df_3QR = df["기말"].quantile(0.75)  # 3사분위수
df_1QR = df_3QR - df_10R            # 사분위 범위
print(df_10R)
print(df_3QR)
print(df_1QR)
```
