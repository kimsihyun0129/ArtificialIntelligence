# 데이터 분석 및 학습 과정

## 1. 필요 패키지 임포트(import)

```python
import numpy as np  # 넘파이 패키지 임포트
import pandas as pd # 판다스 패키지 임포트
import sklearn      # 싸이킷런 패키지 임포트

# 의사결정 나무 분류 모델을 위한 패키지 임포트
from sklearn.tree import DecisionTreeClassifier

# 학습 및 테스트 데이터셋 분리를 위한 패키지 임포트
from sklearn.model_selection import train_test_split
```

## 2. 데이터 수집/불러오기

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

## 3. 데이터 탐색

```python
df.head()
```

```python
df.shape # 데이터프레임의 열과 행의 수를 출력
```

```python
df.info() # 데이터프레임의 요약정보 출력
```

```python
df.describe() # 데이터프레임의 기술통계 보여주기
```

## 4. 데이터 전처리

```python
# 텍스트로 되어 있는 species 컬럼의 데이터를 0,1,2로 변환한다.
df['species'].replace({'setosa':0,'versicolor':1,'virginica':2},inplace=True)
df
```

## 5. 학습 데이터셋 준비

```python
# 분석 데이터셋 준비
# X는 독립변수(설명변수), y는 종속변수(목표변수)
X = df[["sepal_length","sepal_width","petal_length","petal_width"]]
y = df["species"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
print(X_train.shape)  # 학습 데이터(독립변수)
print(X_test.shape)   # 테스트 데이터(독립변수)
print(y_train.shape)  # 학습 데이터(종속변수)
print(y_test.shape)   # 테스트 데이터(종속변수)
```

## 6. 학습 수행

```python
# DecisionTreeClassifier 객체 생성
dt = DecisionTreeClassifier(random_state=11)
dt.fit(X_train, y_train)  # 학습 수행
```

```python
# 학습이 완료된 dt 객체에서 테스트 데이터셋으로 예측 수행
pred = dt.predict(X_test)
```

```python
print(y_test.values)
print(pred)
```

## 7. 성능평가 및 시각화

```python
# 모델 성능 - 정확도 측정
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

### 학습 모델의 성능평가

```python
# 1. 오차 행렬(Confusion Matrix, 혼동 행렬)
# 모델 성능 - 오차 행렬
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```python
# 2. 모델 평가 지표
# 모델 성능 평가 - 평가 지표 계산
from sklearn.metrics import classification_report
rpt = classification_report(y_test, pred)
print(rpt)

# 1 정확도
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)

# 2 정밀도
from sklearn.metrics import precision_score
pre = precision_score(y_test, pred, average=None)
print(pre)

# 3 재현율
from sklearn.metrics import recall_score
rec = recall_score(y_test, pred, average=None)
print(rec)

# 4 F1 Score
from sklearn.metrics import f1_score
f1 = f1_score(y_test, pred, average=None)
print(f1)
```
